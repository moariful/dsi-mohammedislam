```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Read in billboard.csv as a data frame
bd_original = pd.read_csv("../assets/billboard.csv")
```

```python
# The simplest pivot table must have a dataframe and an 
# index. Letâs use the artist as our index.
pd.pivot_table(bd_original, index='artist.inverted')
```

```python
# Let's look at the data by artist and track. 
df2 = pd.pivot_table(bd_original, index=['artist.inverted', 'track'])
```

```python
df2.head()
```

```python
# find out the number of weeks each song has been on the billboard
df3 = pd.melt(bd_original, id_vars=['artist.inverted', 'track'], value_vars=list(bd_original.iloc[:, 7:]))
df3
```

```python
df4 = pd.pivot_table(df3, index='track', aggfunc='count')
```
